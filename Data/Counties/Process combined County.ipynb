```python


#Add Dependencies
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
import numpy as np
```

```python
county_demo = "combined_csv.csv"
file_output = "../2019_County_data.csv"
```

```python
# Import our input dataset
county_df = pd.read_csv(county_demo , dtype=str, low_memory=False, index_col=False, encoding="ISO-8859-1")
```

```python
county_df
```

```python
county_df = county_df.loc[county_df['YEAR'] == "12"]
```

```python
columns_demo = ["STATE", "COUNTY", "STNAME", "CTYNAME", "POPESTIMATE", "POPEST_MALE", "POPEST_FEM", "MEDIAN_AGE_TOT", "MEDIAN_AGE_MALE", "MEDIAN_AGE_FEM" ]
```

```python
countydemo_df=county_df[columns_demo]
```

```python
countydemo_df.info()
```

```python
countydemo_df["FIPS"]= countydemo_df['STATE'] + countydemo_df['COUNTY']
```

```python
#Export dataset to CSV
countydemo_df.to_csv(file_output)
```
